Explanation: This notebook is used to plot out decision tree stages into circular charts.
See examples below:

```python
import matplotlib.pyplot as plt
import numpy as np

labels_parent = ["dress or skirt", "tshirt or shirt or jacket", "pant or short"]
values_parent = [0.69097900390625, 0.10018539428710938, 0.20883941650390625]


labels_stage1 = ["Pencil Skirt", "A-Line Skirt", "Pleated Skirt", "Maxi Skirt", "Wrap Skirt", "A-Line Dress", "Sheath Dress", "Wrap Dress", "Maxi Dress", "Shift Dress"]
values_stage1 = [0.04113578796386719, 0.040866851806640625, 0.0206298828125, 0.009122848510742188, 0.590057373046875, 0.07446670532226562, 0.03625297546386719, 0.033858299255371094, 0.003525257110595703, 0.15007400512695312]


# Function to plot radar chart
def plot_radar_chart(labels, values, color, ax, title=None):
    num_vars = len(labels)
    angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
    values += values[:1]  # Close the loop
    angles += angles[:1]  # Close the loop
    
    ax.set_theta_offset(np.pi / 2)
    ax.set_theta_direction(-1)
    
    ax.plot(angles, values, color=color, linewidth=2, linestyle='solid')
    ax.fill(angles, values, color=color, alpha=0.25)
    ax.set_xticks(angles[:-1])
    ax.set_xticklabels(labels, size=8)
    ax.set_title(title, size=10, pad=20)

# Create the figure
fig, axs = plt.subplots(2, 1, figsize=(10, 10), subplot_kw=dict(polar=True))

# Plot parent stage
plot_radar_chart(labels_parent, values_parent, 'blue', axs[0], title='Parent Avg')

# Plot stage-1-avg part 1
plot_radar_chart(labels_stage1, values_stage1, 'green', axs[1], title='Stage-1-Avg')


# Adjust layout
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data to be visualized in radar chart
labels = ['dress', 'skirt', 'short', 'pant', 'jacket', 'sweatshirt', 'shirt', 'poloshirt', 't-shirt']
values = [0.393218994140625, 0.30194091796875, 0.021183013916015625, 0.07209300994873047, 0.004148960113525391, 0.02790999412536621, 0.0887603759765625, 0.012979507446289062, 0.07773208618164062]


# Calculate number of variables
num_vars = len(labels)

# Compute angle for each category on radar chart
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

# Repeat values to close the radar chart loop
values += values[:1]
angles += angles[:1]

# Initialize radar chart
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Draw one axis per variable and add labels
ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

plt.xticks(angles[:-1], labels)

# Plot data and fill area
ax.plot(angles, values, color='blue', linewidth=2, linestyle='solid')
ax.fill(angles, values, color='blue', alpha=0.25)

# Set y-axis limits
plt.yticks(color="grey", size=7)
plt.title('Prediction Values for Clothing Types (plain-avg)')

# Display the plot
plt.show()
```
